```python
diff = final['difference_charge']
c_charge = final['expected charges as per x']

def output(x, y):
    count_c = 0
    count_u = 0
    count_o = 0

    sum_c = 0
    sum_u = 0
    sum_o = 0
    
    for i, j in zip(x, y):
        if i == 0:
            count_c += 1
            sum_c += j
        elif i < 0:
            count_u += 1
            sum_u += abs(i)
        elif i > 0:
            count_o += 1
            sum_o += i
    
    return (count_c, count_o, count_u), (sum_c, sum_o, sum_u)

a, b = output(diff, c_charge)
print(a, b)
```

```python
output = ['Total orders where x has been correctly charged' , 'Total orders where x has been overcharged',
         'Total orders where x has been undercharged']
df = pd.DataFrame({
    'output' : output ,
    'count' : a,
    'Amount(RS)' : b
})
```
